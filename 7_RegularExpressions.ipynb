#The findall() Function
The findall() function returns a list containing all matches.

```python
# Import library

import re
```

```python
# Print a list of all matches:

txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```python
str1='The match is going on in spain, match'
re.findall('match',str1)
```

```python
str1='The match2 is going on in spain, match2'
re.findall('match',str1)
```

```python
# Return an empty list if no match was found:

txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x)
```

```python
# Search for all the characters [a,ei,o,u] in the string
str1='The match2 is going on in spain, match2'
re.findall('[aeiou]',str1)
```

```python
# Returns a list of charachters other than given in square brackets
str1='The match2 is going on in spain, match2'
print(re.findall('[^aeiou]',str1))
```

```python
# It will check whether the string starts with 't' or not
str1='The match2 is going on in spain, match2'
print(re.findall('^[t]',str1))
```

```python
# It will check whether the string starts with 'hello' or not
text="hello planet, welcome,helooooooo, hi"
re.findall("^hello",text)
```

```python
# It will find exact 2 letters which is between "he" and "o"
text=" hello planet, welcome"
re.findall("he..o",text)
```

```python
text=" helllo planet, welcome"
re.findall("he..o",text)
```

```python
# '*' stands for N number of characters
text=" helllo planet, welcme"
re.findall("he.*o",text)
```

```python
text=" helllo planet, welcome,helooooooo, hi"
re.findall("he.*o",text)
```

```python
# It will check whether the string ends with 'hi' or not
text="hello planet, welcome,helooooooo, hi"
re.findall("hi$",text)
```

```python
# It will check whether the string starts with 'i' or not
text="hello planet, welcome,helooooooo, hi"
re.findall("i$",text)
```

```python
# It will check whether the given text has any digit(0-9)
text='Hi, welcome to python2,285'
re.findall('\d',text)
```

#The search() Function
The search() function searches the string for a match, and returns a Match object if there is a match.

If there is more than one match, only the first occurrence of the match will be returned:

```python
# Search for the first white-space character in the string:

txt = "The rain in Spain"
x = re.search("\s", txt)

print("The first white-space character is located in position:", x.start())
```

```python
#Make a search that returns no match:

import re

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x)
```

#Match Object
A Match Object is an object containing information about the search and the result.

Note: If there is no match, the value None will be returned, instead of the Match Object.

```python
#Do a search that will return a Match Object:

import re

txt = "The rain in Spain"
x = re.search("ai", txt)
print(x) #this will print an object
```

#The Match object has properties and methods used to retrieve information about the search, and the result:

1. span() returns a tuple containing the start-, and end positions of the match.
2. string returns the string passed into the function
3. group() returns the part of the string where there was a match

```python
# Print the position (start- and end-position) of the first match occurrence.
# The regular expression looks for any words that starts with an upper case "S":
import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

#Step-by-step:
###The regex r"\bS\w+" means:
#1. \b → word boundary (start of a word)
#2. S → uppercase letter "S"
#3. \w+ → one or more word characters

###So it matches: "Spain"

```python
#Print the string passed into the function:

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.string)
```

```python
# Print the part of the string where there was a match.
# The regular expression looks for any words that starts with an upper case "S":

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.group())
```

#The split() Function
The split() function returns a list where the string has been split at each match:

```python
# Split at each white-space character:

txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```

```python
# Split the string only at the first occurrence:

txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x)
```

```python
text="the rain in spain"
re.split(' ',text)
```

```python
text="the rain in spain"
re.split(' ',text,1)
```

```python
text="the rain in spain"
re.split(' ',text,4)
```

#The sub() Function
The sub() function replaces the matches with the text of your choice:

```python
#Replace every white-space character with the number 9:

import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt)
print(x)
```

```python
#Replace the first 2 occurrences:

import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2)
print(x)
```
