# Exception Handling:
Exception handling in Python is a control mechanism used to manage runtime errors (exceptions) so that your program doesn’t crash abruptly and can respond gracefully.

```
# Example 1:
try:
    x = int(input("Enter a number: "))
    print(10 / x)
except ZeroDivisionError: # if x = 0
    print("Cannot divide by zero!")
except ValueError: # if x = string
    print("Invalid input! Please enter a number.")
```

```
# Example 3:
try:
    x = int("5")
    y = 10 / x
except ZeroDivisionError:
    print("Division by zero")
else:
    print("Success:", y)
finally:
    print("Execution completed")
```

```
try:
     i=int(input("enter i:"))#wrong input
     j=int(input("enter j:"))
     k=i/j # 9/0 div by zero

except Exception as msg :
   print ("Exception raised:",msg)

else:
  print("Answer:",k)

finally:
  print("the End....")

print("normal end of file and code if any")
```

# Catching Multiple Exceptions

```
try:
    x = int("abc")
except (ValueError, TypeError):
    print("Error occurred")
```

#Catching All Exceptions (Generic)

```
try:
    x = 10 / 0
except Exception as e:
    print("Error:", e)
```

# Raising Exceptions:

You can manually trigger exceptions using "raise".

```
x= int(input("enter number :"))
if x < 18:
  raise Exception("Sorry, You are not eligible for voting, register once you become 18 years old")
else:
  print("you entered :",x)
```

```
def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18 or above")
    return True
```

```
check_age(18)
```

#Common Built-in Exceptions:
1. ValueError – wrong value type
2. TypeError – wrong data type
3. ZeroDivisionError – division by zero
4. IndexError – invalid index
5. KeyError – missing dictionary key
6. FileNotFoundError – file not found
