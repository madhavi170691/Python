#String:

It is a sequence of character enclosed either within single quotes or double quotes

```python
s = "python"
m = 'java'
print(s)
print(type(s))
print(m)
print(type(m))
```

```python
# How to access element?
s = 'python'
print(s)

# index
print(s[2])
print(s[-2])
print(s[-6])
print(s[7])
```

```python
# slice

print(s[::])
print(s[:])
print(s[:4])
print(s[3:]) #hon
print(s[::-1])  #nohtyp
print(s[::-2]) #nhy
print(s[-3:-5:-1])
```

```python
# Mathematical operators
# 1. '+' concatenation
# Rule: both operands should be string

print("python"+10)
```

```python
print("python"+'10')
```

```python
print("python"+' is amazing')
```

```python
# 2. '*' repetition operator
# Rule: one operand should be int and another should be string

print("java " * 2)
print(2 * "java ")
```

```python
# Comparison operator
s1 = 'python'
s2 = 'python'
s3 = 'Python'

print(s1 == s2) # T
print(s1 == s3) # F
print(s1 != s3) # T
print(s1 > s3) # T
print(s1 < s3) # F
```

```python
# Built-in functions

s = 'python programming is amazing'

# len(): it will count the total number of letters present in the given string including space
print(len(s))

```

```python
s = '  python programming is amazing  '
# strip()

print("strip:",s.strip()) # remove whitespaces from both the ends
print("lstrip:",s.lstrip()) # remove whitespaces from left end
print("rstrip:",s.rstrip()) # remove whitespaces from right end
```

```python
s = 'python programming is amazing'
# index(), find() : both returns the index position

print(s.index('p'))
print(s.index('python'))
print(s.index('programming'))
print(s.index('p',1,8))
print(s.index('b'))
```

```python
s = 'python programming is amazing'
print(s.find('p'))
print(s.find('python'))
print(s.find('programming'))
print(s.find('p',1,8))
print(s.find('b'))
```

```python
s = 'python programming'

# count()
print(s.count('p')) #2
print(s.count('python')) #1
```

```python
s = 'python programming'
s1 = s.split() # return type of split function is LIST
print(s1)
```

```python
s = '26-03-2026'
s2 = s.split('-')
print(s2)
```

```python
s = ["varun", "tarun", "arun"]

s1 = ':'.join(s)
print(s1)
```

```python
s = 'Python Program'

print(s.lower())
print(s.upper())
print(s.swapcase())
print(s.title())
print(s.capitalize())
```

```python
s = 'Python Program 123'

print(s.isalnum()) #F
print("python10".isalnum()) #T
print(s.isalpha()) #F
print(s.isdigit()) #F
print(s.islower()) #F
print(s.isupper()) #F
print(s.istitle()) #T

```

```python
l = [10,20,50,60,40,10,10]

print(len(l))
print(l.count(10))
print(l.index(10))
print(min(l))
print(max(l))
```

```python
l = [10,20,50,60,40,10,10]
l.sort()
print(l)

l.sort(reverse=True)
print(l)

```
