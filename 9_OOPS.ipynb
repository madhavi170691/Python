```python
class Human:
  name = 'sadiya'      #(public : reflect, access)
  _age = 35            #(private: not reflect, access)
  __password = "12345" #(strong private: not reflect, not access)
```

```python
class Human:
  name = 'sadiya'      #(public : reflect, access)
  _age = 35            #(private: not reflect, access)
  __password = "12345" #(strong private: not reflect, not access)

  def access(self):
    print(Human.name, Human._age,Human.__password)
```

```python
f = Human()
```

```python
f.name
```

```python
f.access()
```

# Encapsulation:
It combines necessary data + unnecessary data

```python
class BankAccount:
  def __init__(self,balance):
    self.__balance = balance # strong private

  def deposit(self,amount):
    self.__balance += amount

  def withdraw(self,amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Insufficient balance")

  def get_balance(self):
    return self.__balance
```

```python
acc = BankAccount(10000)
```

```python
acc.deposit(5000)
```

```python
acc.withdraw(2000)
```

```python
acc.get_balance()
```

# Inheritance:
A new class(Sub class, derived class) can inherit variables and methods of the existing class(Base class, Super class)

###Types:
1. Single level
2. Multi level
3. Hierarchial
4. Multiple

```python
class Animal:
  def __init__(self,name):
    self.name = name

  def speak(self):
    return "generic animal sound"
```

```python
class Dog(Animal):
  def bark(self):
    return "woof"
```

```python
class Cat(Animal):
  def meow(self):
    return "meow"
```

```python
animal = Animal("generic")
dog = Dog("Tuffy")
cat = Cat("Jack")
```

```python
print(animal.name)
print(animal.speak())
```

```python
print(dog.name)
print(dog.speak())
print(dog.bark())
```

```python
print(cat.name)
print(cat.speak())
print(cat.meow())
```

# Polymorphism:

### Method overloading:

### Method overriding:
It will override the properties and behaviour of parent class.

```python
### Method overloading:
#same function name but different parameter
class Operations:
  def add(self,a,b):
    return a+b
  def add(self,a,b,c):
    return a+b+c

# Python does not support METHOD OVERLOADING bcoz here there is no declaration of datatypes.
```

```python
op = Operations()
```

```python
op.add(10,20)
```

```python
op.add(10,20,30)
```
