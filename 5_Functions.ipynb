```python
def largest(n1,n2,n3):
  if n1 > n2 and n1 > n3:
    print(n1,"is largest")
  elif n2 > n1 and n2 > n3:
    print(n2,"is largest")
  else:
    print(n3,"is largest")
```

```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter third number: "))

largest(n1,n2,n3)
```

```python
def fact(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n*fact(n-1)
```

```python
n = int(input("Enter a number: "))
fact(n)
```
