# Data Structure:

This is nothing but a way to ORGANISE and STORE data so that it can be easily ACCESSED, MANAGED and MODIFIED.

1. List
2. Tuple
3. Set
4. Dictionary

# LIST:
1. It is represented in '[]'.
2. Insertion order is preserved in list.
3. Heterogenous elements are allowed.
4. Duplicates values are also allowed.
5. It is MUTABLE (that can be change).


```python
# Creation of list
l1 = [10]
print(l1)
print(type(l1))
```

```python
l2 = [10,20,3.14,'python',True,10,20]
print(l2)
print(type(l2))
```

```python
# How to access element in list?
l3 = [10,20,3.14,'python',True,10,20]
print(l3[0])
print(l3[2])
print(l3[-3])
```

```python
# Slicing
l4 = [10,20,30,40,50]
print(l4[:3])
print(l4[2:])
print(l4[:])
print(l4[::2])
print(l4[::-1]) # reverse
print(l4[::-2])
print(l4[-4:-2])
print(l4[-2:-4])
```

```python
l5 = [10,20,30,40,50,60]

# Add element
# 1. append(): it always add the element at the end of the list
l5.append(70)
print(l5)

l5.append('java')
print(l5)

# 2. insert(): it will add element at the given index
l5.insert(2,'python')
print(l5)

# 3. extend(): it will add multiple elements

l6 = [10,20,30]
l7 = [40,50]

l6.extend(l7)
print(l6)

l7.extend(l6)
print(l7)

l7.extend('python')
print(l7)
```

```python
l8 = [10,20,30,40,50,60]

#remove(element)
l8.remove(30)
print(l8)

#pop(index)
l8.pop(2)
print(l8)

l8.pop() # it will remove the last element
print(l8)

#Replace element
l8[1] = 'python' # mutable
print(l8)
```

# TUPLE:
1. It is represented in '()'(optional).
2. Insertion order is preserved in tuple.
3. Heterogenous elements are allowed.
4. Duplicates values are also allowed.
5. It is IMMUTABLE (that cannot be changed).

### It is called as READ ONLY VERSION OF LIST.


```python
# Creation of Tuple
t1 = (10,)
print(t1)
print(type(t1))

t2 = 10,20,3.14,'python',True,10,20
print(t2)
print(type(t2))
```

```python
t3 = (10)   # t3 = 10
print(t3)
print(type(t3))
```

```python
t4 = (20,20,30,3.14,50,10)
t4
t4.append('python')
```

# SET: It is known for in UNIQUENESS
1. It is represented in '{}'.
2. Insertion order is not preserved in set.
3. Heterogenous elements are allowed.
4. Duplicates values are not also allowed.
5. It is MUTABLE (that can be changed).


```python
# Creation of sets
s1 = {10}
print(s1)
print(type(s1))

s2 = {10,20,30,3.14,True,10,20,30}
print(s2)
print(type(s2))
```

```python
s3 = {10,20,30,40,50,10}
print(s3)

s3.add(60)
print(s3)

s3.update([70,80,90])
print(s3)

s3.update('python')
print(s3)
```

```python
s4 = {10,20,30,40,50,60}
print(s4)

s4.remove(40)
print(s4)

# s4.remove(70)
# print(s4)

s4.pop()
print(s4)

s4.discard(60)
print(s4)

s4.discard(50)
print(s4)
```

# DICTIONARY:
1. It is represented in key:value term '{k:v}'.
2. Insertion order is not preserved in set.
3. Heterogenous elements are allowed.
4. Duplicates values are allowed but KEYS must be unique.
5. It is MUTABLE (that can be changed).


```python
# creation of dictionary
d = {1:"python",'b':300,3:300 }
print(d)
print(type(d))
```

```python
# How to access element:
print(d[1])
print(d['b'])
print(d[3])
```

```python
# How to update elemnts
d = {1:"A",2:"B",3:"C"}
print(d)

d[4] = "D"
print(d)

d[4] = "E"
print(d)
```

```python
d = {1:"A",2:"B",3:"C"}
print(d)
```

```python
del d[3]
print(d)
```

```python
d.clear()
print(d)
```

```python
del d
```

```python
d = {1:"A",2:"B",3:"C"}
print(d)

print(d.keys())
print(d.values())
print(d.items())
```
