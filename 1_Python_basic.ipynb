# Identifier:

A name given to any class, function or variable.


### Allowed Characters:
1. All alphabets (A-Z, a-z)
2. All digits (0-9)
3. Special Characters (only underscore allowed '_')


### Rules to define identifier:
1. It can starts with Alphabets (A-Z, a-z). --> eg: Total = 50, x = 1000, Total_9 = 10
2. It can starts with '_' but it will act as a Private variable. --> eg: _total = 20
3. It is Case sensitive. --> eg: Total = 10 is different from TOTAL = 10
4. It cannot starts with digit. --> eg: 9_Total = 500
5. It cannot use KEYWORDS. --> eg: if, for, pass, break, continue.
6. No length limit. --> eg: Total = 10, Tttttoootaallll = 10
7. It cannot use Dollar sign '$'.  --> $total = 30, Total$ = 300

# Operators

```python
# Arithmetic operator(+,-,*,/,//,%,**):

a = 10
b = 2

print("Addition :",a+b)
print("Difference :",a-b)
print("Mulplication: ",a*b)
print("Division: ",a/b)
print("Floor Division: ",a//b)
print("Remainder: ",a%b)
print("Power: ",a**b)
```

```python
# Comparison Operator (==,!=,>,>=,<,<=)

a = 10
b = 10
c = 20

print(a==b) #T
print(a==c) #F
print(a!=b) #F
print(a!=c) #T
print(c>a)  #T
print(c>=a) #T
print(b<c)  #T
print(a<=b) #T
```
