# 6. Line Chart (Trend analysis)

# 7. Scatter Plot

```python
plt.scatter(df['Sales'], df['Profit'], color = 'magenta')
plt.title('Profit vs Sales')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# Using Seaborn (Statistical & Advanced) :
Best for insightful, visually rich analysis

# 8. Bar Plot (Aggregated)

```python
import seaborn as sns

sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Sales by Region')
plt.xticks(rotation=90)
plt.show()
```

# 9. Heatmap (correlation)

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# 10. Pairplot (multivariate)

```python
sns.pairplot(df[['Sales', 'Profit', 'Inventory', 'Returns']])
plt.show()
```

# 11. CountPlot

```python
sns.countplot(x='Product', data=df, color = 'red')
plt.title('Product Count')
plt.xticks(rotation=60)
plt.show()
```

```python
sns.countplot(x='Customer_Segment', hue='Region', data=df)
plt.title('Customer Segment Distribution by Region')
plt.show()
```
